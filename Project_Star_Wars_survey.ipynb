```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

print(star_wars.columns)
star_wars.head(2)
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head()

#or
# star_wars.dropna(subset = ['RespndentID], inplace = True)
```

```python
yes_no = {'Yes':True, 'No':False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
import numpy as np
dict_seen = {"Star Wars: Episode I  The Phantom Menace": True, 
             "Star Wars: Episode II  Attack of the Clones":True,
             "Star Wars: Episode III  Revenge of the Sith":True,
             "Star Wars: Episode IV  A New Hope":True,
             "Star Wars: Episode V The Empire Strikes Back":True,
             "Star Wars: Episode VI Return of the Jedi":True,
             np.NaN: False
            }
columns = star_wars.columns[3:9]


for col in columns:
    star_wars[col] = star_wars[col].map(dict_seen)

star_wars.head()
```

```python
star_wars.rename(columns = {"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
                  "Unnamed: 4" : 'seen_2',
                  "Unnamed: 5" : 'seen_3',
                  "Unnamed: 6" : 'seen_4',
                  "Unnamed: 7" : 'seen_5',
                  "Unnamed: 8" : 'seen_6'}, inplace = True)
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.rename(columns = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_movie1',
                           'Unnamed: 10' : 'ranking_movie2',
                           'Unnamed: 11' : 'ranking_movie3',
                           'Unnamed: 12' : 'ranking_movie4',
                           'Unnamed: 13' : 'ranking_movie5',
                           'Unnamed: 14' : 'ranking_movie6',}, inplace = True)
star_wars.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticklabels(['ranking_movie1', 'ranking_movie2', 'ranking_movie3', 'ranking_movie4', 'ranking_movie5', 'ranking_movie6'], rotation = 90)

plt.bar(range(6),star_wars[star_wars.columns[9:15]].mean())
plt.show()

#star_wars.hist(column = star_wars.columns[9:15])

```

```python
seen = star_wars[star_wars.columns[3:9]].sum()
print(seen)
fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xticklabels(['seen_1','seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6'])

plt.bar(range(6), seen)
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax.set_xticklabels(['ranking_movie1', 'ranking_movie2', 'ranking_movie3', 'ranking_movie4', 'ranking_movie5', 'ranking_movie6'], rotation = 90)
ax2.set_xticklabels(['ranking_movie1', 'ranking_movie2', 'ranking_movie3', 'ranking_movie4', 'ranking_movie5', 'ranking_movie6'], rotation = 90)

ax.bar(range(6),males[males.columns[9:15]].mean(), color = 'darkred')
ax2.bar(range(6),females[females.columns[9:15]].mean(), color = 'darkblue')
plt.show()
```

```python
star_wars_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
star_war_not_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax.set_xticklabels(['ranking_movie1', 'ranking_movie2', 'ranking_movie3', 'ranking_movie4', 'ranking_movie5', 'ranking_movie6'], rotation = 90)
ax2.set_xticklabels(['ranking_movie1', 'ranking_movie2', 'ranking_movie3', 'ranking_movie4', 'ranking_movie5', 'ranking_movie6'], rotation = 90)

ax.bar(range(6),star_wars_fan[star_wars_fan.columns[9:15]].mean(), color = 'darkred')
ax2.bar(range(6),star_war_not_fan[star_war_not_fan.columns[9:15]].mean(), color = 'darkblue')
plt.show()
```

```python
star_trek_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'Yes']
star_trek_not_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'No']

import seaborn as sns
sns.barplot(star_trek_fan.columns[9:15], y = star_trek_fan[star_trek_fan.columns[9:15]].mean(), color = 'salmon')
plt.xticks(rotation = 30)
plt.show()
sns.barplot(star_trek_not_fan.columns[9:15], y = star_trek_not_fan[star_trek_not_fan.columns[9:15]].mean(), color = 'green')
plt.xticks(rotation = 30)
plt.show()

```
